```python
import random

# Define a list of product categories
product_categories = [
    "Fruits & Vegetables", "Bakery", "Meat & Seafood", "Dairy",
    "Pantry Staples", "Frozen Food", "Snacks & Beverages", "Specialty Items"
]

# Define a function to generate a random typical purchase
def generate_typical_purchase(categories):
  num_items = random.randint(2, 5)
  purchase = []
  for _ in range(num_items):
    category = random.choice(categories)
    item = f"{category} - {random.randint(1, 5)} items"
    purchase.append(item)
  return ", ".join(purchase)

# Define a function to generate a random customer
def generate_customer():
  age = random.randint(18, 75)
  loyalty_member = random.choice(["Yes", "No"])
  household_size = random.randint(1, 5)
  primary_shopper = random.choice(["Self", "Spouse", "Parents"])

  # Diets can be None or a random choice from the list
  diets = random.choices(["Vegetarian", "Vegan", "Gluten-free", None], weights=[0.1, 0.05, 0.05, 0.8], k=1)[0]

  # Randomly choose spending amount based on household size
  avg_spend = {1:200, 2:300, 3:400, 4:500, 5:600}
  monthly_spend = avg_spend.get(household_size, 300) + random.randint(-50, 50)

  # Random chance of having preferred brands
  preferred_brands = random.choice(["Yes", "No"])

  # Randomly generate visit frequency
  visited_last_week = random.choice(["Yes", "No"])
  visits_per_month = random.randint(1, 4)

  # Generate random email and phone number (not for real use)
  email = f"customer.{random.randint(100,999)}@email.com"
  phone_number = f"(555) 555-{random.randint(1000,9999)}"

  # Define random choices for other attributes
  occupation = random.choice(["Tech Professional", "Student", "Teacher", "Healthcare", "Retail", "Homemaker", "Retired"])
  communication_channel = random.choice(["Email", "SMS", "Both"])
  social_media_following = random.choice(["Y", "N"])

  # Generate interests based on typical purchase
  typical_purchase = generate_typical_purchase(product_categories)
  interests = ", ".join([category.split()[0] for category in typical_purchase.split(", ")])

  # Reward program participation can be Yes or No
  reward_program = random.choice(["Yes", "No"])

  return {
      "ID": random.randint(1, 1000),
      "Name": f"Customer {random.randint(100,999)}",
      "Age": age,
      "Loyalty_Member": loyalty_member,
      "Typical_Purchase": typical_purchase,
      "Preferred_Payment": random.choice(["Credit Card", "Debit Card", "Cash"]),
      "Household_Size": household_size,
      "Primary_Grocery_Shopper": primary_shopper,
      "Diets_Restrictions": diets,
      "Average_Monthly_Spend": monthly_spend,
      "Preferred_Brands_(if_any)": preferred_brands,
      "Visited_Last_Week?": visited_last_week,
      "Visits_per_Month_(Avg)": visits_per_month,
      "Email_Address_(example)": email,
      "Phone_Number_(example)": phone_number,
      "Occupation": occupation,
      "Preferred_Communication_Channel_(Email/SMS/Both)": communication_channel,
      "Social_Media_Following_(Y/N)": social_media_following,
      "Interests_(derived_from_purchase_history)": interests,
      "Reward_Program_Participation": reward_program
  }

# Generate 50 customer entries
customers = [generate_customer() for _ in range(50)]

# Print the data (optional)
```

```python
import random
from datetime import date, timedelta

# Define product categories
product_categories = [
    "Fruits & Vegetables", "Bakery", "Meat & Seafood", "Dairy",
    "Pantry Staples", "Frozen Food", "Snacks & Beverages", "Specialty Items"
]

# Define a function to generate a random product name
def generate_product_name(category):
  if category == "Fruits & Vegetables":
    return f"{random.choice(['Apple', 'Banana', 'Orange', 'Lettuce', 'Potato'])} - {random.randint(1, 5)} lbs"
  elif category == "Bakery":
    return random.choice(["Bread", "Cake", "Cookies", "Muffins", "Pastries"])
  elif category == "Meat & Seafood":
    return f"{random.choice(['Chicken', 'Beef', 'Fish', 'Pork'])} - {random.choice(['Ground', 'Whole', 'Sliced'])} ({random.randint(1, 2)} lbs)"
  elif category == "Dairy":
    return random.choice(["Milk", "Cheese", "Yogurt", "Butter", "Eggs"])
  elif category == "Pantry Staples":
    return random.choice(["Rice", "Pasta", "Cereal", "Beans", "Soup"])
  elif category == "Frozen Food":
    return random.choice(["Pizza", "Ice Cream",  "Vegetables", "Entrees", "Snacks"])
  else:
    return random.choice(["Coffee", "Candy", "Chips", "Spices", "Condiments"])

# Define a function to generate random inventory levels
def generate_inventory_levels():
  stock_level = random.randint(10, 100)
  reorder_point = random.randint(5, stock_level - 5)
  return stock_level, reorder_point

# Define a function to generate a random expiration date (future date)
def generate_expiry_date(days):
  base_date = date.today()
  random_days = random.randint(days, days * 2)
  return base_date + timedelta(days=random_days)

# Define a function to generate a random supplier
def generate_supplier():
  return f"Supplier {random.randint(1, 10)}"

# Define a function to generate a random inventory entry
def generate_inventory_entry():
  category = random.choice(product_categories)
  product_name = generate_product_name(category)
  stock_level, reorder_point = generate_inventory_levels()
  expiry_date = generate_expiry_date(7) if category in ["Dairy", "Bakery"] else None
  supplier = generate_supplier()
  cost_price = random.uniform(1.00, 10.00)
  retail_price = round(cost_price * 1.2, 2)

  return {
      "ID": random.randint(1000, 9999),
      "Product_Category": category,
      "Product_Name": product_name,
      "Stock_Level": stock_level,
      "Reorder_Point": reorder_point,
      "Expiry_Date": expiry_date.strftime("%Y-%m-%d") if expiry_date else None,
      "Supplier": supplier,
      "Cost_Price": cost_price,
      "Retail_Price": retail_price,
      "Location": random.choice(["Aisle 1", "Aisle 2", "Freezer", "Refrigerator"]),
      "Date_Added": date.today().strftime("%Y-%m-%d"),
      "Notes": random.choice(["", "On promotion", "Low stock"]),
      "Average_Sales_per_Week": random.randint(10, 50),
      "Last_Updated": date.today().strftime("%Y-%m-%d"),
      "Image": f"product_{random.randint(1, 100)}.jpg",  # Placeholder image name
      "Active": random.choice([True, False]),  # Simulate inactive items
  }

# Generate 100 inventory entries
inventory = [generate_inventory_entry() for _ in range(100)]

# Print the data (optional)
# for item in inventory:
#   print(item)
```
